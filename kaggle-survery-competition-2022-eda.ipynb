# Kaggle Survey 2022 EDA

This notebook explores the Kaggle Survey 2022 dataset, analyzing participant demographics, skills, tools, and roles across the data science community. We aim to understand gender differences, skills by job title, and broader trends in the field.

## Objectives
- Understand the demographic distribution (e.g., gender, education)
- Explore the most common tools and programming languages
- Compare roles (e.g., Data Scientist, Data Analyst) across skills and tools
- Identify trends and insights that stand out in the survey data

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt  
import plotly.graph_objects as go
import plotly.express as px 

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('/kaggle/input/kaggle-survery-competition-2022/kaggle_survey_2022_responses.csv')
df.shape
```

```python
# Preview the dataset
df.head()
```

```python
# Remove the top row
df = df.iloc[1:,:]
```

## Data Overview
Let's check the dataset summary and null value percentages.

```python
df.describe()
```

```python
# Get percent of null values in question
df.isnull().sum() / df.shape[0]
```

```python
#create a dictionary for questions 
Questions = {}

#create list of questions 
#not very efficient, but keeps things ordered
qnums = list(dict.fromkeys([i.split('_')[0] for i in df.columns]))
qnums
```

```python
#add data for each question to key value pairs in dictionary
for i in qnums:
    if i in ['Q2','Q3']: #since we are using .startswith() below this prevents all questions that start with 
        Questions[i] = df[i] #[1,2,3] from going in the key value pair (Example in vid)
    else:
        Questions[i] = df[[q for q in df.columns if q.startswith(i)]]
```

```python
df['Q2']
```

```python
# Q2 histogram using px 
fig = px.histogram(df, x = 'Q2', category_orders={'Q2':['18-21','22-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-69','70+']})
fig.show()
```

```python
# heatmap using px for q2 & q11
fig = px.density_heatmap(df, x='Q2', y='Q11', category_orders={'Q2':['18-21','22-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-69','70+'],'Q11':['I have never written code','< 1 years','1-2 years','3-5 years','5-10 years','10-20 years','20+ years']})
fig.show()
```

```python
Questions['Q12']
```

```python
# Q7 example for go use. We aggregate the data beforehand with .value_counts()
Questions['Q12'].columns = list(Questions['Q12'].mode().iloc[0,:])
q12 = Questions['Q12'].count().reset_index()
q12.columns = ['language','Count']
q12 = q12.sort_values('Count', ascending = False)
fig = go.Figure([go.Bar(x = q12.language, y = q12.Count)])
fig.show()
```

```python
#Create dictionary with role / data key value pairs
Roles = {}
for i in df['Q23'].unique():
    Roles[i] = df[df['Q23'] == i]
```
